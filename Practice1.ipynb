```python
import numpy as np 
from scipy import stats 
import pandas as pd 
import matplotlib.pyplot as plt 
from scipy.stats import skew 
data = [4,4,5,6,7,1,3,5,7]
print(f"Number of student is --> {len(data)}")
print(f'The median point is -->{np.median(data)}')
print(f'The Mean point is -->{np.mean(data)}')
print(f'The Skewness is -->{skew(data)}')
print(f'The Standard Deviation -->{np.std(data)}')
print(f'The Zscores is -->{stats.zscore(data)}')
print(f'The Tscores is -->{np.tscore(data)}')
```
